### 통합필드 생성

```python
sigun = pd.read_excel("results/ecm_result.4.0.xlsx")
```

```python
do = { 
    "세종특별자치시": "세종",
    "제주특별자치도": "제주",
    "서울특별시": "서울",
    "부산광역시": "부산",
    "울산광역시": "울산",
    "대구광역시": "대구",
    "광주광역시": "광주",
    "대전광역시": "대전"
    }

sigun["시군구 제거"] = sigun["시군구"].str.replace("시$", "", regex=True).str.replace("군$", "", regex=True)
sigun["광역시"] =sigun["시군구"].map(do)

```
