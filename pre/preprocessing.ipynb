```python
nonhouse.drop(index=[0,1,2,3],inplace=True)
nonhouse.head(8)

nonhouse = nonhouse[
    ~(
        (nonhouse['월'].str[-2:] != "12") |
        (nonhouse['월'].str[:4] == "2023") |
        (nonhouse['구분'].isin(['서울', '부산', '대구', '인천', '광주', '대전', '울산'])) &
        (nonhouse['시군구'].str.strip() != "계") 

    )
]

nonhouse.to_excel('pp_nonhouse.xlsx', index=False)
```
