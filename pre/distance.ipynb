```python
import pandas as pd
import numpy as np
import math

def get_distance(lat1, lon1, lat2, lon2):
    R = 6371  # 지구 반지름 (킬로미터)
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])  # 각도를 라디안으로 변환
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    distance = R * c
    
    distance = round(distance)
    return distance


```

```python
df = pd.read_excel('EA_Data/location.xlsx')
```

```python
location = df["시도 시군구 위도 (Latitude) 경도 (Longitude)"].str.split(" ", expand=True)
location.columns = ["시도", "시군구", "위도", "위도방향", "경도","경도방향"]
```

```python
location.drop(columns=["시도","위도방향","경도방향"])
```

```python
location["위도"] = location["위도"].str.replace("°", "", regex=False).astype(float)
location["경도"] = location["경도"].str.replace("°", "", regex=False).astype(float)
```

```python
seoul_lat, seoul_long = 37.5665, 126.9780
```

```python
location["서울과의 거리"] = location.apply(lambda location: get_distance(seoul_lat, seoul_long, location["위도"], location["경도"]), axis=1)
```
